```python
ACTIONS = ["L", "R", "U", "D"]
STATES = [1,2,3,4,5,6,7,8,9]
REWARDS = {1:-0.05, 2:-0.05, 3:-0.05, 4:-0.05, 5:-10, 6:-0.05, 7:15, 8:-0.05, 9:30}
DISCOUNT = 0.99

# PROBFUNCTION = {
#     1 : {
#         "U" : {4: 1},
#         "D" : {1:1},
#         "L" : {1:1},
#         "R" : {2:1}
#     },
#     2 : {
#         "U" : {5:1},
#         "D" : {2:1},
#         "L" : {1:1},
#         "R" : {3:1}
#     },
#     3 : {
#         "U" : {6:1},
#         "D" : {3:1},
#         "L" : {2:1},
#         "R" : {3:1}
#     },
#     4 : {
#         "D" : {1:1},
#         "R" : {5:1},
#         "L" : {4:1},
#         "U" : {7:1}
#     },
#     5 : {
#         "U" : {},
#         "D" : {},
#         "L" : {},
#         "R" : {},
#     },
#     6 : {
#         "U" : {9:1},
#         "D" : {3:1},
#         "L" : {5:1},
#         "R" : {6:1}
#     },
#     7 : {
#         "U" : {},
#         "D" : {},
#         "L" : {},
#         "R" : {}
#     },
#     8 : {
#         "U" : {8:1},
#         "D" : {5:1},
#         "L" : {7:1},
#         "R" : {9:1}
#     },
#     9 : {
#         "U" : {},
#         "D" : {},
#         "L" : {},
#         "R" : {}
#     }
# }

PROBFUNCTION = {
    1 : {
        "U" : {4:0.9, 2:0.1},
        "D" : {1:0.9, 2:0.05, 4:0.05},
        "L" : {1:0.9, 2:0.05,  4:0.05},
        "R" : {2:0.9,  4:0.1}
    },
    2 : {
        "U" : {5 : 0.9, 3:0.05, 1:0.05},
        "D" : {2:0.9, 1:0.1/3, 3:0.1/3, 5:0.1/3},
        "L" : {1: 0.9, 5:0.05, 3:0.05},
        "R" : {3: 0.9, 5:0.05, 1:0.05}
    },
    3 : {
        "U" : {6:0.9, 2:0.1},
        "D" : {3:0.9, 2:0.05, 6:0.05},
        "L" : {2:0.9, 6:0.1},
        "R" : {3:0.9, 2:0.05, 6:0.05}
    },
    4 : {
        "D" : {1:0.9, 7:0.05, 5:0.05},
        "R" : {5:0.9, 7:0.05, 1:0.05},
        "L" : {4:0.9, 7:0.1/3, 5:0.1/3, 1:0.1/3},
        "U" : {7:0.9, 5:0.05, 1:0.05}
    },
    5 : {
        "U" : {},
        "D" : {},
        "L" : {},
        "R" : {},
    },
    6 : {
        "U" : {9:0.9, 3:0.05, 5:0.05},
        "D" : {3:0.9, 9:0.05, 5:0.05},
        "L" : {5:0.9, 9:0.05, 3:0.05},
        "R" : {6:0.9, 9:0.1/3, 3:0.1/3, 5:0.1/3}
    },
    7 : {
        "U" : {},
        "D" : {},
        "L" : {},
        "R" : {}
    },
    8 : {
        "U" : {8:0.9, 5:0.1/3, 7:0.1/3, 9:0.1/3},
        "D" : {5:0.9, 9:0.05, 7:0.05},
        "L" : {7:0.9, 5:0.05, 9:0.05},
        "R" : {9:0.9, 5:0.05, 7:0.05}
    },
    9 : {
        "U" : {},
        "D" : {},
        "L" : {},
        "R" : {}
    }
}

policy_str = ""

# Initialize the value function
value = {i:0 for i in STATES}
policy = {i:"L" for i in STATES}


for i in range(7):
    new_value = {i:0 for i in STATES}
    policy_str += " | ".join([ policy[_s] for _s in STATES]) + "\n"
    print(i, end=" | ")
    [print(f"{value[i]:3.5f}",end=" | ") for i in STATES]
    print()
    for state in STATES:
        best = -float("inf")
        best_action = policy[state]
        for action in ACTIONS:
            temp_sum = [PROBFUNCTION[state][action].get(_s,0)*value[_s] for _s in STATES]
            temp_sum = sum(temp_sum)
            if temp_sum > best:
                best = temp_sum              
                best_action = action
        new_value[state] = REWARDS[state] + DISCOUNT*best
        policy[state] = best_action
    value = new_value
print(7, end=" | ")
[print(f"{value[i]:3.5f}",end=" | ") for i in STATES]
print("\n")
policy_str += " | ".join([policy[i] for i in STATES]) + "\n"
print(policy_str)

```
